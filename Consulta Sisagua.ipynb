```python

import pandas as pd
import pyarrow
import pnadium
import matplotlib
import openpyxl
import numpy
```

```python
df_sisagua = pd.read_csv("vigilancia_parametros_basicos_2021.csv", sep=';', encoding='latin1', low_memory=False)
display(df_sisagua)
display(df_sisagua.shape)
display(df_sisagua.info())
display(df_sisagua.describe())
```

```python
df_sisagua['CO_MUNICIPIO_IBGE'] = df_sisagua['CO_MUNICIPIO_IBGE'].astype(str)
```

```python
df_sisagua[df_sisagua['NU_AMOSTRA'] == '213216000207']
```

```python
amostras_por_municipio = (
    df_sisagua
    .groupby(['CO_MUNICIPIO_IBGE', 'NO_MUNICIPIO'])['NU_AMOSTRA']
    .nunique()
    .reset_index(name='QTDE_AMOSTRAS')
)
display(amostras_por_municipio)
display(amostras_por_municipio.shape)
display(amostras_por_municipio.info())
display(amostras_por_municipio.describe())
```

```python
amostras_por_municipio[amostras_por_municipio['QTDE_AMOSTRAS'] == 4366]
```

```python
amostras_por_municipio[amostras_por_municipio['QTDE_AMOSTRAS'] == 1]
```

```python
df_sisagua[df_sisagua['CO_MUNICIPIO_IBGE'] == '241360']
```

```python
amostras_parametro_municipio = (df_sisagua
                                .groupby(['CO_MUNICIPIO_IBGE', 'NO_MUNICIPIO', 'PARAMETRO'])['NU_AMOSTRA']
                                .nunique()
                                .reset_index(name='QTDE_AMOSTRAS'))
display(amostras_parametro_municipio)
display(amostras_parametro_municipio.shape)
display(amostras_parametro_municipio.info())
display(amostras_parametro_municipio.describe())
```

```python
amostras_parametro_municipio.to_excel("amostras_parametro_municipio.xlsx",index=False)
```
